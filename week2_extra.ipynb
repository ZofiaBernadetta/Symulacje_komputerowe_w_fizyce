Wykonać wykres (pikselowy) przecięcia Poincare dla możliwie długiej
ewolucji układu Duffinga w obszarze chaotycznym.
Zaznaczać punkty (xn,vn) otrzymane w dyskretnych chwilach czasu,
odpowiadających pełnemu okresowi siły pobudzającej tn =
2π
/ω
*n (dla
naturalnego n). Powstaje w ten sposób portret dziwnego atraktora.
Przykładowe parametry: c=0.05, f=0.3.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

f = 0.3
ω = 2*np.pi*0.213

def duffing(t, xv):
    x, v = xv
    a, b, c = 1, 1, 0.05
    return [v, b*x - (x**3 * a ) - (c*v) + (f * np.cos(ω*t)) ]


a, b = 1, 100000
t = np.linspace(a, b, 100000) * ( 2 * np.pi / ω )
print(t)


sol1 = solve_ivp(duffing, [a, t[-1]], [1,1], t_eval=t)

plt.scatter(sol1.y[0][50:], sol1.y[1][50:], s=3, c='g', lw=0, marker='o', label='punkty')
plt.xlim(right=3)  # adjust the right leaving left unchanged
plt.xlim(left=-3)
plt.ylim(top=3)  # adjust the right leaving left unchanged
plt.ylim(bottom=-3)
plt.xlabel("$x$")
plt.ylabel("$v$")
plt.savefig("lorenz_xt.png")
plt.show()



```
