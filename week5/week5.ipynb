```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import imageio.v2 as imageio
import os

os.makedirs('images', exist_ok=True)

N = 16
box_size = 8.0
eps = 1.0
sigma = 1.0
particle_radius = sigma / 2
dt = 0.0001
temp = 2.5
kB = 1
m = 1

dt = 0.001

class particle:

    rs = np.array([])
    p_halfs = np.array([])

    ts = np.array([0])
    v_halfs = np.array([])

    def __init__(self, particle_radius, r, v, id):
        self.particle_radius = particle_radius
        self.r = r
        self.v = v
        self.id = id
        self.rs = np.array([r])
        self.vs = np.array([v])



particles = []

for i in range(4):
    for j in range(4):

        v = np.array([np.random.uniform(-0.5, 0.5), np.random.uniform(-0.5, 0.5)])
        # make dependent on box size
        r = np.array([1 + 2*i, 1 + 2*j])
        p = particle(particle_radius, r, v, f'{i}{j}')
        particles.append(p)


v = np.array([p.v for p in particles])
v_cm = sum(v) / N

for particle in particles:
    particle.v -= v_cm

def kinetic_energy(particle):
    return 0.5 * m * particle.r ** 2

Ek_mean = sum([kinetic_energy(p) for p in particles])

fs = np.sqrt(kB * temp / Ek_mean)

for particle in particles:
    particle.v = fs * particle.v

def unit_vector(r):
    x, y = r
    magnitude = np.sqrt(x**2 + y**2)
    if magnitude == 0:
        raise ValueError("Zero vector has no direction — cannot compute unit vector.")
    return r / magnitude

def forceFrom(r):
    magnitude = np.linalg.norm(r)
    if magnitude == 0:
        return [0.,0.]
    return -48 * eps * unit_vector(r) *( (( sigma / magnitude ) ** 13) - (0.5 * (( sigma / magnitude ) ** 7) )) / sigma

def force(i, rs, particles_rs):
    force = np.array([0., 0.])
    for ri in particles_rs:

    # trzeba przeskalować do tego kawdratu
        distance =  rs[-1] - ri[-1]
        distance = closest_picture(distance)
       # print(r)
        force += forceFrom(distance)
    return force


filenames = []
def closest_picture(r):
    if r[0] > box_size/2:
        r[0] -= box_size
    if r[0] < -box_size/2:
        r[0] += box_size
    if r[1] > box_size/2:
        r[1] -= box_size
    if r[1] < -box_size/2:
        r[1] += box_size
    return r

def move_back_to_box(r):
    if r[0] > box_size:
        r[0] -= box_size
    if r[0] < 0:
        r[0] += box_size
    if r[1] > box_size:
        r[1] -= box_size
    if r[1] < 0:
        r[1] += box_size
    return r

def trajectory(particles, steps=100000):

    #r = np.array([r0])
    #p_half = np.array([p0 - force(r0) * dt / 2])
    #p = np.array([p0])
    #t = np.array([0])

#self.v_halfs = np.array([v - force(r) * dt / 2])

    for i in range(steps):
        for particle in particles:

            F = force(i, particle.rs, [p.rs for p in particles])

            vi_half = particle.v_halfs[-1] if len(particle.v_halfs) != 0 else  particle.vs[0] - ( F * dt / (2 * m) )

            ri = particle.rs[-1]

            ri = move_back_to_box(ri)


            vii_half = vi_half + (F * dt / m)
            #print(f'vii_half = {vii_half} for {i}')

            rii = move_back_to_box(ri + vii_half * dt)

# zrobić z tego funkcję


            #vii = vii_half - F * dt / ( 2 * m )
            print(particle.rs.shape)
            particle.rs = np.append(particle.rs, [rii], axis=0)
            print(particle.rs.shape)
            particle.v_halfs = np.vstack([particle.v_halfs, vii_half]) if len(particle.v_halfs) > 0 else np.array([vii_half])
            #particle.vs = np.append(particle.vs, [vii], axis=0)
            particle.ts = np.append(particle.ts, particle.ts[-1] + dt)
        if i % 100 == 0:
            plot(i)
    return particles

def plot(i):
    plt.clf() # wyczyść obrazek
    fig = plt.gcf() # zdefiniuj nowy
    for p in particles: # pętla po cząstkach
        a = plt.gca()
        cir = Circle((p.rs[-1][0], p.rs[-1][1]), radius = p.particle_radius) # kółko tam gdzie jest cząst
        a.add_patch(cir) # dodaj to kółko do rysunku
        plt.plot() # narysuj
    plt.xlim((0, box_size)) # obszar do narysowania
    plt.ylim((0, box_size))
    fig.set_size_inches((6, 6)) # rozmiar rysunku
    plt.title(f'Symulacja gazu Lennarda-Jonesa, krok {i:06d}')
    name = f'images/img{i:06d}.png'
    filenames.append(name)
    plt.savefig(name)

trajectory(particles)

with imageio.get_writer('images/movie.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)


```
