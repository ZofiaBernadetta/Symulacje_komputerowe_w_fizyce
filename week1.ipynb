Wygeneruj tablicę o wymiarach N x N (np. dla N = 10, 50, 100), w
komórkach której znajdą się liczby wylosowane z przedziału [0,1] zgodnie
z jednostajnym rozkładem prawdopodobieństwa.
Następnie przygotuj wykres, który będzie graficznie przedstawiał
wygenerowaną tablicę (przydatne może okazać się plt.imshow). Do
kolorowania wykresu wybierz wybraną przez siebie mapę barw.
Dalej spraw, aby wiersze tablicy przesunęły się o jeden w dół (tzn. 1.
wiersz stał się drugim, 2. wiersz stał się trzecim itd, a wiersz o numerze N
stanie się pierwszym, przyda się funkcja np.roll).
Wygeneruj gifa, który będzie prezentował cały cykl przesuwania się
wierszy (tj. wykonuj operację opisaną powyżej do momentu, gdy wiersz
który zaczynał jako pierwszy ponownie zostanie pierwszym).

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v3 as iio


n = 100

a = np.random.rand(n,n)
images = [ ]

for i in range(n):
    plt.imshow(a, cmap='viridis', interpolation='nearest')
    plt.savefig(f'{i}.png')
    a = np.roll(a, 1, axis=0)
    images.append(iio.imread(f'{i}.png'))

iio.imwrite('team.gif', images, duration = 500, loop = 0)

```
